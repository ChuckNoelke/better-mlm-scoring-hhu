```python
import pandas as pd

df = pd.read_csv("clean_EventsAdapt_SentenceSet.csv")
sentences = list(df["Sentence"])
words = [sent.lower().rstrip(".").split() for sent in sentences]
words = [item for sublist in words for item in sublist]
words = sorted(list(set(words)))
words
```

```python
with open('EventsAdapt_vocabulary.txt', 'w') as f:
    for word in words:
        f.write(f"{word}\n")
```
