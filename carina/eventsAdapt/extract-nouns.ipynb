```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn
nouns = {x.name().split('.', 1)[0] for x in wn.all_synsets('n')}
```

```python
with open("EventsAdapt_vocabulary.txt") as file:
    words = [line.rstrip() for line in file]
```

```python
events_nouns = [elm for elm in words if elm in nouns]
print(f"Found {len(events_nouns)} nouns in the whole of the EventsAdapt vocab of size {len(words)}!")
```

```python
events_nouns
```

```python
with open('EventsAdapt_vocabulary_nounsOnly.txt', 'w') as f:
    for noun in events_nouns:
        f.write(f"{noun}\n")
```
