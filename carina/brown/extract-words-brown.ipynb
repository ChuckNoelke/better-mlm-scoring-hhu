```python
import pickle

with open('brown_stimuli.pkl', 'rb') as f:
    df = pickle.load(f)
sentences = df["sentence"].values
words = [sent.lower().rstrip(".").split() for sent in sentences]
words = [item for sublist in words for item in sublist]
words = [word for word in words if word.isalpha()] #keep only letter strings
words = sorted(list(set(words)))
words
```

```python
with open('Brown_vocabulary.txt', 'w') as f:
    for word in words:
        f.write(f"{word}\n")
```

# Split into sublists

```python
import numpy as np
vocab_chunks = np.array_split(words,10)
for ind, chunk in enumerate(vocab_chunks):
    print(len(chunk))
    with open(f'Brown_vocabulary_chunk_{ind}.txt', 'w') as f:
        for word in chunk:
            f.write(f"{word}\n")
```
