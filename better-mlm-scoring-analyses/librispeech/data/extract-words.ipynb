```python
import json
```

```python
sentences = []
with open('test-clean.am.json') as json_file:
    corpus = json.load(json_file)
    for sent_idx, value in corpus.items():
        ref_stimulus = value["ref"] # Only scoring ref sentences so far
        sentences.append(ref_stimulus)
```

```python
words = [sent.lower().rstrip(".").split() for sent in sentences]
words = [item for sublist in words for item in sublist]
words = sorted(list(set(words)))
words
```

```python
with open('LibriSpeech_vocabulary.txt', 'w') as f:
    for word in words:
        f.write(f"{word}\n")
```
