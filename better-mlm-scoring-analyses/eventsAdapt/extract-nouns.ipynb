```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet as wn
nouns = {x.name().split('.', 1)[0] for x in wn.all_synsets('n')}
```

```python
with open("EventsAdapt_vocabulary.txt") as file:
    words = [line.rstrip() for line in file]
```

```python
events_nouns_nltk = [elm for elm in words if elm in nouns]
print(f"NLTK found {len(events_nouns_nltk)} nouns in the whole of the EventsAdapt vocab of size {len(words)}!")
```

```python
events_nouns_nltk
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

events_nouns_spacy = [elm for elm in words if nlp(elm)[0].tag_.startswith("NN")]
print(f"SPACY found {len(events_nouns_spacy)} nouns in the whole of the EventsAdapt vocab of size {len(words)}!")
```

```python
events_nouns_spacy
```

```python
for elm in events_nouns_spacy:
    if elm not in events_nouns_nltk:
        print(elm)
print("*****")
for elm in events_nouns_nltk:
    if elm not in events_nouns_spacy:
        print(elm)
```

```python
nouns = sorted(list(set(events_nouns_nltk).union(events_nouns_spacy)))
print(f"Overall we found {len(nouns)} nouns in the whole of the EventsAdapt vocab of size {len(words)}!")
```

```python
with open('EventsAdapt_vocabulary_nounsOnly.txt', 'w') as f:
    for noun in nouns:
        f.write(f"{noun}\n")
```
